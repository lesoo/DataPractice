# Data Preparation
- train, test split -> row 분리
- 분제(feature)와 정답(label) 분리 -> column 분리(데이터의 방향성에 맞게)
- 날짜, 날씨, 시간과 같은 정보가 casual, registered, count에 영향을 줄것


```python
df.columns
```

### feature, label split

```python
features=['season', 'holiday', 'workingday', 'weather', 'temp', 'atemp',
       'humidity', 'windspeed', 'casual', 'registered', 'count', 'year',
       'month', 'day', 'hour', 'dayofweek']
#문제
```

```python
label='count'
#답
```

### train, test split

```python
train, test=df[0::2], df[1::2]
#짝수->훈련, 홀수->테스트
```

```python
test
```

```python
train, test=train.reset_index(), test.reset_index()
#숫자 재정렬
```

```python
test
```

```python
x_train, y_train=train[features], train[label]
#문제->x, 답->y
x_test, y_test=test[features], test[label]
```

```python
x_test
```

```python
y_test
```

# Modeling

- 지도학습의 사례 2가지(객관식, 주관식)
- Classification(분류) : 객관식(경우의 수)
    > 사진->답
    > 사진->뭐같음?
- Regression(회귀) : 주관식(숫자)
    > 예측(연속적인 값 ex)가격, 주가, 매출, 환율, 부동산)

### 지금은 회귀


```python
from sklearn.ensemble import RandomForestRegressor as rf
#scikit-learn

#앙상블 기법
#RendomFrorest
```

```python
rf()
```

```python
model=rf()
#알고리즘 생성

model.fit(x_train, y_train)
#fit-학습시킨다
#문제를 주고 학습하는 개념
```

```python
#시험보고 점수 확인
test_score=model.score(x_test, y_test)
test_score
```

#같은 데이터셋으로도 수치가 미미하게 다를수 있음

# 과대적합과 과소적합
- overfitting - 샘플에 모델이 과하게 적합되어있는 상태
- underfitting - 샘플에 모델이 너무 부족하게 적합되어있는 상태

# 결정계수(R-squared, R2)
- 회귀분석이 얼마나 잘되었는지 평가하는 지표
R^2=1-(sse/sst)=1-(target-예측값)^2/(target-org)^2
- 샘플의 타깃과 예측한 값의 차이를 제곱하여 더함
- 타깃과 타깃평균의 차이를 제곱하여 더한 값으로 나눔
- 타깃의 평균 정도를 예측하는 수준이라면 (분자와 분모가 비슷하다면) R2는 0에 가까워지고
- 예측이 타깃에 가까워지면(분자가 0에 가까워지기 때문에) 1에 가까운 값이 됨
