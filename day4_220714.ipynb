# 데이터분석 절차, 단계, 방법론
- KDD, CRISP-DM


## CRISP-DM
- Cross Industry Standard Process - Data Mining
- 가장 많이 사용되는 Data Mining 표준 방법론
- 분야, 산업군을 막론하고 데이터로부터 의미를 도출하는 표준 절차

# Business Understanding
- 분석할 분야의 목적과 요구사항을 이해하고, 초기 프로젝트 계획 수립단계
- 업무 목적과 상황을 파악, 프로젝트 계획 수립 등의 세부 업무 포함
- 해당 분야 종사자(클라이언트)가 원하는 방향 캐치
- 분야별 관심사가 다름 
    - ex)제조업 - 재구매율, 생산량, 매출액, 불량률 등등
    - ex2)숙박업 - 재방문률, 방문율, 예약 취소율
    - ex3)학원 - 환불확률, 취소율
- 과거 학술지, 기술적으로 선행연구가 무엇이었는지 파악, 자문 구하기
- 분석 방향성 및 목표 설정


## Data Understanding
- 데이터 수집, 데이터 속성 이해하는 단계
- 가용 데이터로 기존에 설정한 분석 방향성과 목표 달성이 가능할지 여부 확인
### EDA(Exploratory Data Analysis)
- 시각화를 통해 데이터 안의 유효한 패턴(insight)을 확인 가능하게 만듦
- 수집한 데이터를 관찰하고 이해하는 과정
    1. 데이터의 출처와 주제에 대한 이해
    2. 데이터의 크기 확인(데이터의 대략적인 양이나 개수 확인)<br>
        ->sampling (전수조사x)
    3. 데이터의 구성요소(feature) 확인(column=feature)(데이터간 상호역할 확인)
    4. 탐색한 데이터의 시각화<br>
        -> 결측치 빈도, 일관성, 형태(자료형) 확인 ex)표기의 통일성), 
- 데이터 기술분석, 데이터 품질 확인 등이 있음


## Data Preparation
- 분석 기법에 적합한 데이터를 편성하는 단계, 데이터셋 선택, 데이터 정제 등
- 모델링(Machine Learning)을 대비하는 단계
- 알고리즘에 넣을 데이터를 '적절하게' 준비

### Machine Learning
- 지도 학습 - 문제과 정답(label)이 존재, 이를 대응하게 알려주는것(강아지 사진 -> 강아지, 0는 0다)
- 비지도 학습 - 정답이 없이 문제만 부여, 답이 없는 문제, 데이터를 보고 스스로 구분
- 학습용(train)과 테스트용 데이터 분리(split)<br>
    ->필수 작업 - 학습했던 문제와 시험문제가 동일하면 시험점수 신뢰 불가
- 데이터셋 분리는 학습용 데이터가 테스트에 나오게 하지 않기 위함
<br><br>
** 과적합(overfitting) - 학습용 데이터가 과다해 테스트에 실패한 경우 -> '적절하게'
