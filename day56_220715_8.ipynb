# 과대적합과 과소적합
- overfitting - 샘플에 모델이 과하게 적합되어있는 상태
- underfitting - 샘플에 모델이 너무 부족하게 적합되어있는 상태

# 결정계수(R-squared, R2)
- 회귀분석이 얼마나 잘되었는지 평가하는 지표
R^2=1-(sse/sst)=1-(target-예측값)^2/(target-org)^2
- 샘플의 타깃과 예측한 값의 차이를 제곱하여 더함
- 타깃과 타깃평균의 차이를 제곱하여 더한 값으로 나눔
- 타깃의 평균 정도를 예측하는 수준이라면 (분자와 분모가 비슷하다면) R2는 0에 가까워지고
- 예측이 타깃에 가까워지면(분자가 0에 가까워지기 때문에) 1에 가까운 값이 됨

### 다양한 회귀문제 평가지표들(약어 파악 중요)(여유있을때 따로 공부)
- R(Root), M(Mean), S(Square), L(Log), P(Percentile), A(Absolute), E(Error)
- MSE(Mean Squared Error) - 오차를 제곱해서 평균내기
- MAE - 오차를 절대값 씌우고 평균
- MAPE - 오차의 비율을 절대값 씌우고 평균
- RMSE - 오차를 제곱하여 평균한 것에 제곱근 적용
- RMSLE - 로그오차를 제곱하여 평균한 것에 제곱근 적용
    > 작은 값들끼리의 차이와 큰 값들끼리의 차이에 차등을 두기 위함
    > 로그 왜씀? - ex) 문제:답 - 1:2, 1001:1002 의 차이

# Feature selection(변수 고르기)(차원축소와 다름)
- 과적합 방지, 일반화 달성 위함
- 모든 변수 사용x, 일부 골라서 사용(영향도 고려)
- 학습속도 증진 기대 가능
> 방법론 3가지 : filter, wrapper, emded


## filter
- 기초통계량 기반, feature간의 높은 상관계수를 가지는 feature를 사용하는 방법
- ex)센서류의 경우 종종 발생(표준편차가 0에 가까운 feature)
- filter방식이 유효한 경우는 분야에 따라 다름

```python
df.describe().T.sort_values(by='std')
```

## wrapper(시험 많이)
- feature의 조합에 따라 평가지표가 달라진다 에서 생김
- feature의 조합별로 평가지표 점수를 확인하여 점수가 가장 좋은 조합을 찾는것
- 전진선택(forward selection), 후방제거(backward elimination), 단계별 확인(stepwise selection)

#### features에서 칼럼을 하나씩 지우고 더해가며 점수를 확인하고, 중요도를 확인

```python
from sklearn.feature_selection import RFE
#feature 제거해주는 거
```

```python
features=['season', 'holiday', 'workingday', 'weather', 'temp', 'atemp',
       'humidity', 'windspeed', 'year',
       'month', 'day', 'hour', 'dayofweek']
label=['casual', 'registered', 'count']
train, test=df[0::2], df[1::2]
#짝수->훈련, 홀수->테스트
x_train, y_train=train[features], train[label]
#문제->x, 답->y
x_test, y_test=test[features], test[label]
train, test=train.reset_index(), test.reset_index()
#숫자 재정렬
model=rf()
#알고리즘 생성

model.fit(x_train, y_train)
#fit-학습시킨다
#문제를 주고 학습하는 개념

#여기에 추가!!!
selector=RFE(model)
selector.fit(x_train, y_train)

model.score(x_test, y_test)
```

```python

selector.ranking_
#중요도에 따라 순위매김, 상위권 feature를 삭제하면 점수 변동폭 크다 
```

```python
rfe_df=pd.DataFrame()
rfe_df['rank']=selector.ranking_
rfe_df['feature']=features
rfe_df.sort_values(by='rank')
#랭킹을 맹신하면 안된다
```
