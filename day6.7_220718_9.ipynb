# Group 연산
- 특정 기준으로 그룹을 분할시켜 처리하는 방식
- 그룹 연산은 데이터를 집계, 변환, 필터링 하는데에 효율적

```python
df.groupby('year')
#이게 그룹화 시킨거임
```

```python
df.groupby('year').mean()
```

```python
year=df.groupby('year')
#year기준 row그룹지음
year_count=year.agg({'count':np.mean})
#agg->그룹별 합산연산
#{count:np.mean}->count에 대해 np.mean연산 일괄수행
#연도별 count의 평균
year_count.reset_index()
```

```python
df.groupby('year').agg({'count':np.mean}).reset_index()
#위에꺼랑 같은얘기
```

```python
#연도별, 계절별 이용량 평균
#2011 여름 이용량
df.groupby(['year', 'season']).agg({'count':np.mean}).reset_index
```
